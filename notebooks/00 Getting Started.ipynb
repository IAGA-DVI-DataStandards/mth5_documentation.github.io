# Before You Begin With MTH5

There are a few things to check on your computer before you start playing with MTH5. In particular, you will need some method to set up Anaconda environments. On Linux/Mac, this is as easy as running an installer command such as ``apt-get`` or ``brew``. On Windows, we recommend downloading [Anaconda Prompt](https://docs.anaconda.com/anaconda/install/windows/) or [Miniforge](https://github.com/conda-forge/miniforge) and working in that console throughout. To this end, we will assume you have a relatively up-to-date version of Python installed as well.

If your machine does not include Git (namely if you're on Windows), you may download an interactive bash console for it [here](https://gitforwindows.org/). This may be useful if you wish to download code directly and edit it for specific tasks.

And with our prerequisites installed, let's get mth5 set up. First, we create a conda environment via:

```python
conda create -n mtenv
```

Next, activate the environment:

```python
conda activate mtenv
```

Then installing mth5 is as simple as running one of the following commands:

```python
pip install mth5
pip install git+https://github.com/kujaku11/mth5.git
```

If you desire to have an interactive installation of the code for implementing your own features, you can also use the following commands:

```python
git clone https://github.com/kujaku11/mth5.git
cd mth5
pip install -e .
```

We also strongly recommend having a local version of ``jupyter-lab`` so you can run code inside of a notebook environment and keep everything organized. To do this, simply run:

```python
pip install jupyter-lab
```

Then you can initialize a JupyterLab kernel by executing:

```python
jupyter notebook
```
