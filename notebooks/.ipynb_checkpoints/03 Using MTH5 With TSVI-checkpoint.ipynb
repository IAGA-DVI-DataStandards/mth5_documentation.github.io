# Using MTH5 With TSVI

One of the more practical uses of MTH5 is being able to visualize and plot different datasets read into it from raw data streams. We can use the timeseries viewing TSVI library to help with this. TSVI provides an inutitive, easy-to-use interface for reading and plotting multiple runs and channels of data within in MTH5 file. This notebook will step you through how to do this and show off a few of the main features as well.

The code for TSVI can be found readily on GitHub. We recommend using the USGS fork of the code. If you are on Windows and cannot run ``git`` commands natively, you can download Git Bash [here](https://gitforwindows.org/) and then run this command inside the console:

```python
git clone https://github.com/kujaku11/tsvi.git
```

Next, to keep everything organized, we'll create a Python environment to run this all in. On Windows, you can use [Anaconda Prompt](https://docs.anaconda.com/anaconda/install/windows/) or [Miniforge](https://github.com/conda-forge/miniforge) as your console.

```python
conda create -n mtenv python=3.9
```

You can change the Python to a more recent version as you please. Either way, next up is to activate the environment via:

```python
conda activate mtenv
```

Finally, we can install TSVI plus its dependencies by using ``pip``:

```python
cd tsvi
pip install -e .
```

Alternatively, you can install TSVI directly from Git:

```python
pip install git+https://github.com/kujaku11/tsvi.git
```

And with that, TSVI is installed and ready for use (though, as you will see, we're more interested in the prerequisite libraries). You may find it useful to install a few other libraries beforehand as well, which we will do here:

```python
pip install jupyter-lab
pip install git+https://github.com/kujaku11/mth5.git
pip install git+https://github.com/simpeg/aurora.git
```

To start working with TSVI in browser, we'll use ``panel`` to run an interactive window generated by ``mth5_viewer_v2.py``. This will have been installed to your environment by way of the ``pip`` commands from earlier, so from the top level directory inside your ``tsvi`` git clone, you can run:

```python
panel serve tsvi/mth5viewer/mth5_viewer_v2.py --show
```

Wait a few seconds for everything to load up and you will see a tab in your default web browser open:



From here, you can navigate to wherever your mth5 file of interest is via the searchbar or file browser, then click on the two right karats to add it to the data. Let's use the ``from_nims.h5`` that we made in our NIMS tutorial:



Then navigate to the DataFrame tab to select a run of interest. Here you can see the date and time range for a given run, the number of samples in the run, the sample rate, and input/output channels, just as if you were viewing the mth5 through Jupyter notebook:



Then, on the sidebar, simply click Plot and let TSVI do the rest of the work:



You should get a plot of each channel in the run that looks something like this:



The default settings checkboxes on the sidebar should suffice for most users. ``Calibrate`` removes sensor and instrument response from the data and generates data measured in physical units instead of mV, which need to be converted to electric and magnetic field appropriate units in postprocessing. ``Subtract Mean`` subtracts the average reading over a given channel from all of its data points before plotting, useful if you wish to see how the graph oscillates above and below said average reading.
